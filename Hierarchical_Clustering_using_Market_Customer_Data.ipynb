```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Genre'] = le.fit_transform(df['Genre'])
```

```python
x = df.drop(['CustomerID', 'Genre','Age'], axis=1)
```

```python
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt
dendogram = shc.dendrogram(shc.linkage(x, method='ward'))
plt.title('Dendogram')
```

```python
from sklearn.cluster import AgglomerativeClustering
ac = AgglomerativeClustering(n_clusters=5,metric='euclidean',  linkage = 'average')
ac.fit(x)
```

```python
df['Cluster'] = ac.fit_predict(x)
```

```python
import seaborn as sns
plt.figure(figsize=(10, 8))
sns.scatterplot(x='Annual_Income_(k$)', y='Spending_Score', hue='Cluster', data=df, palette='viridis', s=100, alpha=0.8)
plt.title('Agglomerative Clustering')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend(title='Cluster')
plt.show()
```
